## 3. UPLOAD, READ EXCEL & DEFINE SHEET DATA LOG PSM

### 3.1 Read and Added sheet Master Data Lookup

```python
df_dataLookup = pd.read_excel('PSMAnalysis_TemplateProcessing.xlsx', sheet_name='master data lookup')
```

```python
df_dataLookup.head()
```

### 3.2 Added Column CLUSTER (ECOMMERCE) on Data Log PSM by VLOOKUP MASTER DATA LOOKUP

```python
df = pd.merge(left=df, right=df_dataLookup, how='left',
        left_on='Banner Description', right_on='Key Cust Level 4  Banner Desc')
```

```python
df.head()
```

### 3.3 Delete Column

```python
df = df.drop(['Key Cust Level 4  Banner Desc'], axis=1)
```

```python
df.head() #column Key Cust Level 4  Banner Desc was deleted
```

### 3.4 Added Column LOSS ALLOC

```python
df['Loss Alloc'] = df['Order Quantity'] - df['New Confirmed Quantity']
```

```python
df.head(20)
```

### 3.5 Added Column Unicode

```python
df['unicode'] = df[['Material', 'Plant']].apply(lambda x: ''.join(x.dropna().astype(str)), axis=1)
```

```python
df.head(50)
```

### 3.6 Move Unicode Column to Left

```python
move_unicode = df.pop(df.columns[-1])
```

```python
df.insert(0, move_unicode.name, move_unicode)
```

### 3.7 Filter Cluster Column just E-COMMERCE Data

```python
filterCluster = df.loc[df['Cluster'] == 'E-COMMERCE']
```

```python
pd.options.display.max_columns = None
filterCluster.head()
```

###  3.8 Update & Save on New Excel with Sheet Define (data log PSM)

```python
filterCluster.to_excel('24D5_B2_Data E-Comm Log Report PSM.xlsx', sheet_name='data log PSM', index=False)
```

## 4. READ & DEFINE ANDROMEDA SHEET

```python
df_stock = pd.read_excel('PSMAnalysis_TemplateProcessing.xlsx', sheet_name='ANDROMEDA')
```

```python
df_stock.head()
```

### 4.1 Added Column Unicode

```python
df_stock['unicode'] = df_stock[['Material', 'Plant']].apply(lambda x: ''.join(x.dropna().astype(str)), axis=1)
```

```python
df_stock.head()
```

### 4.2 Move Unicode Column to Left

```python
move_unicode = df_stock.pop(df_stock.columns[-1]) #get data unicode
```

```python
df_stock.insert(0, move_unicode.name, move_unicode) #paste data into left
```

###  4.3 Save Data to Excel on ANDROMEDA sheet name

```python
with pd.ExcelWriter('24D5_B2_Data E-Comm Log Report PSM.xlsx', engine='openpyxl', mode='a') as writer:
     df_stock.to_excel(writer, sheet_name='ANDROMEDA', index=False)
```

```python
df_stock.head() #sudah ter-save di excel
```

## 5. FINAL SHEET

### 5.1 Define Excel & Data Log PSM Sheet

```python
xls = pd.ExcelFile('24D5_B2_Data E-Comm Log Report PSM.xlsx')
logPSM = xls.parse('data log PSM')
```

```python
logPSM.head()
```

### 5.2 Define Excel & ANDROMEDA Sheet

```python
andromeda = xls.parse('ANDROMEDA')
```

```python
andromeda.head()
```

### 5.3 Lookup by Unicode

```python
df_final = pd.merge(left=logPSM, right=andromeda, how='left', 
        left_on='unicode', right_on='unicode')
```

```python
pd.options.display.max_columns = None
df_final.head()
```

### 5.4 Delete Columns Material_y & Plant_y

```python
df_final.drop(['Material_y', 'Plant_y'], axis=1)
df_final.head()
```

### 5.5 Rename Columns Material_x = Material & Plant_x = Plant

```python
df_final.rename(columns={'Material_x':'Material', 'Plant_x':'Plant'}, inplace=True)
df_final.head()
```

### 5.6 Save Dataframe to FINAL2 Sheet

```python
with pd.ExcelWriter('24D5_B2_Data E-Comm Log Report PSM.xlsx', engine='openpyxl', mode='a') as writer:
     df_final.to_excel(writer, sheet_name='FINAL2', index=False)
```

### 5.7 Pivot to FINAL Sheet by FINAL2 Sheet

```python
final_data = ['unicode', 'Material', 'Material Description', 'Sold to', 'Sold To Description', 'New Goods Issue date', 'New RDD', 'Sales Document', 'Purchase Order Number', 'Plant', 'Cluster', 'HANA Log', 'Loss Alloc', 'Available', 'QI Stock', 'IT This wk', 'IT Next wk']
pd.options.display.max_columns = None
df_final[final_data].head()
```

### 5.8 Save to FINAL Sheet

```python
with pd.ExcelWriter('24D5_B2_Data E-Comm Log Report PSM.xlsx', engine='openpyxl', mode='a') as writer:
     df_final[final_data].to_excel(writer, sheet_name='FINAL', index=False)
```

### 5.9 Delete FINAL2 Sheet

```python
filePSM = openpyxl.load_workbook('24D5_B2_Data E-Comm Log Report PSM.xlsx')
```

```python
filePSM.remove(filePSM['FINAL2'])
```

### 5.10 Save Excel

```python
filePSM.save('24D5_B2_Data E-Comm Log Report PSM.xlsx')
```
